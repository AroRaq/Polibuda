```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Simulation parameters
T = 5
h = 0.1

#Model parameters
alpha = 2
x_0 = 3
```

```python
t = np.arange(0, T, h)
x = np.zeros(t.shape)
x[0] = x_0

for i in range(t.size-1):
    x[i+1] = x[i] + h*alpha*x[i]

plt.plot(t, x)
```
