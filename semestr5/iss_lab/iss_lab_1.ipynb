```python
import random
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def find_zero(loss_function, left, right, e=0.01, stop_after=1000):
    X, Y = [], []
    no_progress_iters = 0
    best_x, best_y = -1, 1000000
    while no_progress_iters < stop_after:
        x = random.uniform(left, right)
        y = loss_function(x)
        X.append(x)
        Y.append(y)
        if y < best_y:
            if best_y - y > e:
                no_progress_iters = 0
            best_x, best_y = x, y
        no_progress_iters = no_progress_iters + 1

    return X, Y, best_x, best_y
```

```python
def plot_result(loss_function, left, right, best_x, best_y, X, Y):
    t1 = np.arange(left, right, (right-left)/100)
    plt.plot(t1, loss_function(t1), X, Y, 'r+', best_x, best_y, 'wo')
```

```python
def f(x):
    return -np.sin(x) * x
```

```python
X, Y, best_x, best_y = find_zero(f, -10, 10)
plot_result(f, -10, 10, best_x, best_y, X, Y)
```
